## Training y Testing

 En esta etapa dividiremos el set de datos en training y testing.

Crearemos el conjunto de Features (X) y Labels (Y)


```python
features = dataset[['Attr40','Attr2','Attr3','Attr9']]
labels = dataset['Attr59']
```

Importamos la librería de sklearn que nos ayudará a dividir el set de datos.

```python
from sklearn.model_selection import train_test_split
```

Para nuestro split la combinación que utilizaremos es de 70% de Training y 30% de Testing

```python
X_train,X_test, y_train, y_test = train_test_split(features, labels,test_size=0.30, random_state = 42)
```

Revisaremos el tamaño de nuestro set de Datos

A continuación aplicaremos un clasificador. Para mi proyecto aplicaré Decission Tree. La naturaleza de mi proyecto es de clasificación por lo que me parece el modelo mas adecuado a utilizar.

```python
from sklearn.tree import DecisionTreeClassifier
```

Definiré a continuación la profundida Máxima de mi árbol:

```python
DecisionTreeModel = DecisionTreeClassifier(max_depth = 3)
```

 A continuación aplicamos el entrenamiento:

```python
DecisionTreeModel.fit(X_train,y_train)
```

 La siguiente métrica me permitirá evaluar el accuracy de mi módelo y que tan bien generaliza:

```python
from sklearn.metrics import accuracy_score
```

```python
y_predict = DecisionTreeModel.predict(X_test)
```

```python
Accu_resultado = accuracy_score(y_test, y_predict)
Accu_resultado 
```
