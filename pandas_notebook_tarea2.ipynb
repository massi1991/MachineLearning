# Datos Faltantes

##### Vamos a ver como podemos eliminar aquellos rows en los que nos falta al menos un valor en alguna columna.
##### Vamos a eliminar los rows que no tengan los datos completos. En este ejemplo usamos all. 
##### Esto significa que va a eliminar la fila o columna solo si todos los datos para esa fila o columna son NaN

```python
dataset.dropna(how='all')
```

##### Vamos a eliminar los rows que no tengan los datos completos. En este ejemplo usamos ANY. 
##### Esto significa que va a eliminar la fila  solo si al menos uno de los datos para esa fila es NaN.

```python
dataset.dropna( how='any')
```

##### Vamos a eliminar las columnas que no tengan los datos completos. En este ejemplo usamos ANY. 
##### Esto significa que va a eliminar la columna solo si al menos uno de los datos para esa columna es NaN. Usaremos Axis 1 que es para columnas.

```python
dataset.dropna(axis=1, how='any')
```

##### Rellenar con un valor los NaN

```python
dataset.fillna(value=0)
```
